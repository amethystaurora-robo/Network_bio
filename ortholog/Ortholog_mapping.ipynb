```python
import pandas as pd
import numpy as np
import re
```

```python
fruit_fly = pd.read_csv('magna_melanogaster.tsv',sep='\t')
human = pd.read_csv('dma_hsa.tsv',sep='\t')
daphnia_pulex = pd.read_csv('dma_dpx.tsv',sep='\t')
dosed_genes = pd.read_csv('../omics/dosed_genes.csv')
protein_info = pd.read_csv('enzyme_proteins.csv')
kegg_terms = pd.read_csv('../GRN/DynGENIE3/user_ko.tsv',sep="\t")
```

```python
#create a file which combines all 3 ortholog mappings in 4 columns - magna, pulex, fruit fly, human, titled human_daphnia
fly_human = pd.merge(fruit_fly, human, on='Daphnia_magna')
human_daphnia = pd.merge(fly_human, daphnia_pulex, on='Daphnia_magna')
```

```python
human_daphnia.to_csv('all_orthologs.csv')
```

```python
#clean Dapma column in KEGG terms file
kegg_terms['regulatory'] = kegg_terms['regulatory'].str.replace('t1$', '', regex=True)
#merge protein information with KEGG terms
kegg_protein = pd.merge(kegg_terms,protein_info,on='KO')
print(len(kegg_protein['KO'].unique()))
print(kegg_protein['KO'].unique())
kegg_protein.head()
kegg_protein.to_csv('protein_info_kegg_terms.csv')
```

```python
#merge protein information with differentially expressed genes
proteins_dosed_genes = pd.merge(kegg_protein,dosed_genes,on='regulatory')
print(len(proteins_dosed_genes))
print(proteins_dosed_genes['KO'].notna().sum())
print(proteins_dosed_genes['KO'].unique())
proteins_dosed_genes.head()
```
